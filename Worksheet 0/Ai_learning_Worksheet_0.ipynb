```python
def convert_units(conversion_type, value):
    """
    Converts a value between different units.

    Parameters:
    conversion_type (int): The type of conversion to perform.
    value (float): The value to be converted.
    """

    if conversion_type == 1:  # meters to feet
        return value * 3.28084
    elif conversion_type == 2:  # feet to meters
        return value / 3.28084
    elif conversion_type == 3:  # kilograms to pounds
        return value * 2.20462
    elif conversion_type == 4:  # pounds to kilograms
        return value / 2.20462
    elif conversion_type == 5:  # liters to gallons
        return value * 0.264172
    elif conversion_type == 6:  # gallons to liters
        return value / 0.264172
    else:
        return None


def main():
    print("Unit Converter")
    print("1. Meters to Feet")
    print("2. Feet to Meters")
    print("3. Kilograms to Pounds")
    print("4. Pounds to Kilograms")
    print("5. Liters to Gallons")
    print("6. Gallons to Liters")

    try:
        choice = int(input("Choose conversion type (1-6): "))

        if choice < 1 or choice > 6:
            print("Unsupported conversion type.")
            return

        value = float(input("Enter the value to convert: "))

        result = convert_units(choice, value)

        if result is not None:
            print("Converted value:", round(result, 3))
        else:
            print("Conversion failed.")

    except ValueError:
        print("Invalid input! Please enter numeric values.")


# Run the program
main()
```

```python
# Function to calculate sum
def find_sum(numbers):
    """
    Returns the sum of a list of numbers.

    """
    return sum(numbers)


# Function to calculate average
def find_average(numbers):
    """
    Returns the average of a list of numbers.

    """
    return sum(numbers) / len(numbers)


# Function to find maximum
def find_maximum(numbers):
    """
    Returns the maximum value from a list of numbers.

    """
    return max(numbers)


# Function to find minimum
def find_minimum(numbers):
    """
    Returns the minimum value from a list of numbers.

    """
    return min(numbers)


def main():
    print("Mathematical Operations on a List")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    try:
        choice = int(input("Choose an operation (1-4): "))

        numbers_input = input("Enter numbers separated by spaces: ")
        numbers = [float(num) for num in numbers_input.split()]

        if len(numbers) == 0:
            print("Error: The list cannot be empty.")
            return

        if choice == 1:
            print("Sum:", find_sum(numbers))
        elif choice == 2:
            print("Average:", find_average(numbers))
        elif choice == 3:
            print("Maximum:", find_maximum(numbers))
        elif choice == 4:
            print("Minimum:", find_minimum(numbers))
        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Invalid input! Please enter numeric values only.")


# Run program
main()
```

```python
#  Problem - 4: Matrix Operations + Linear Equations
print("=== Problem 4: Matrix Operations ===")

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. A * A_inv = I
A_inv = np.linalg.inv(A)
print("1. A @ A_inv:\n", np.round(A @ A_inv, decimals=10))
print("   (Should be Identity)")

# 2. AB ≠ BA
AB = A @ B
BA = B @ A
print("\n2. AB:\n", AB)
print("   BA:\n", BA)
print("   AB == BA?", np.array_equal(AB, BA))

# 3. (AB)^T == B^T @ A^T
print("\n3. (AB).T:\n", AB.T)
print("   B.T @ A.T:\n", B.T @ A.T)
print("   Equal?", np.array_equal(AB.T, B.T @ A.T))

# Linear Equation System
print("\n=== Solving Linear System ===")
coeff = np.array([[2, -3, 1],
                  [1, -1, 2],
                  [3,  1, -1]])
const = np.array([-1, -3, 9])

# Using inverse
inv_coeff = np.linalg.inv(coeff)
solution = inv_coeff @ const
print("Solution using np.linalg.inv:")
print("x =", solution[0])
print("y =", solution[1])
print("z =", solution[2])

# Bonus: Using np.linalg.solve (faster & recommended)
solution2 = np.linalg.solve(coeff, const)
print("\nSame solution with np.linalg.solve:", solution2)
```
